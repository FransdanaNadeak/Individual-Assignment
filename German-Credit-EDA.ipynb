```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
data=pd.read_csv('german_credit_data.csv')
data.head()
```

```python
data=data.drop(data.columns[0],axis=1)
data.head()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(data.isnull(), yticklabels=False, cbar=False)
```

```python
#count how much the percentage of missing value for each field
percentage = round(data.isnull().sum()/len(data)*100,2).sort_values(ascending = False)
percentage
```

```python
f=plt.figure(figsize=(15,4))
f.add_subplot(1,3,1)
sns.distplot(data['Age'])
f.add_subplot(1,3,2)
sns.distplot(data['Credit amount'])
f.add_subplot(1,3,3)
sns.distplot(data['Duration'])
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
#create correlation with hitmap
Numerical=['Age','Duration','Credit amount']
#create correlation
corr = data[Numerical].corr(method = 'pearson')

#convert correlation to numpy array
mask = np.array(corr)

#to mask the repetitive value for each pair
mask[np.tril_indices_from(mask)] = False
fig, ax = plt.subplots(figsize = (15,12))
fig.set_size_inches(10,5)
sns.heatmap(corr, 
            mask = mask, 
            vmax = 0.9, 
            square = True, 
            annot = True)
```
