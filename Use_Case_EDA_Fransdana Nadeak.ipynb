### 3.  Identify missing value. If any, how will you handle it? 

```python
#count how many missing values per field
count_missing = data.isnull().sum().sort_values(ascending = False)
count_missing
```

Dari hasil pengecekan missing value, terdapat 7 Variabel yg memiliki missing value dengan rincian, 6 Variabel Numerik dan 1 Variabel kategorik.

Variabel Numerik akan dimputasi dengan Mean, sedangkan Variabel Kategorik akan diimputasi dengan Modus

#### Imputasi Variabel Numerik dengan Mean

```python
data.fillna(data.mean(), inplace=True)
```

#### Imputasi Variabel Kategorik dengan Modus

```python
data['num-of-doors'].fillna(data['num-of-doors'].mode()[0], inplace=True)
```

```python
#count how many missing values per field
count_missing = data.isnull().sum().sort_values(ascending = False)
count_missing
```

Setelah dilakukan imputasi, sudah tidak didapatkan Variabel yang mengandung missing value

## 4.  Perform visualization using at least 5 difference visualization technique

### Numerical Variable Visualization Using Histogram and Violinplot

```python
f=plt.figure(figsize=(15,10))

f.add_subplot(2,2,1)
plt.style.use('ggplot')
plt.hist(data['price'],
         facecolor='blue',
         edgecolor='yellow',
         bins=10)
plt.title('Histogram of Price')

#plt.show()
f.add_subplot(2,2,2)
plt.hist(data['stroke'],
         facecolor='blue',
         edgecolor='yellow',
         bins=10)
plt.title('Histogram of stroke')

f.add_subplot(2,2,3)
plt.violinplot(data['horsepower'])
plt.title('Violin Plot of horsepower')

f.add_subplot(2,2,4)
plt.violinplot(data['length'])
plt.title('Violin Plot of length')

plt.show()


```

### Barplot for Categorical Variable

```python
import seaborn as sns
make=pd.DataFrame(data['make'].value_counts())
make=make.reset_index()
make=make.sort_values('make')
make

plt.figure(figsize=(10,5))
plt.barh(make['index'],make['make'])
plt.show()
```

### Scatter Plot to see the relation between two numerical variables

```python
f=plt.figure(figsize=(10,5))
f.add_subplot(1,2,1)
plt.scatter(data['width'],data['price'])
plt.title('Width VS Price')

f.add_subplot(1,2,2)
plt.scatter(data['length'],data['price'])
plt.title('Length VS Price')

plt.show()
```

### Pie Chart to know the proportion of Categorical Variable

```python
# Pie chart
body_style = ['sedan', 'hatchback', 'wagon', 'hardtop','convertible']
sizes = [96, 70, 25, 8, 6]
#colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','green']
 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=body_style, autopct='%1.1f%%', startangle=100)
#draw circle
centre_circle = plt.Circle((0,0),0.7,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

## Correlation Plot

```python
#data.info()
numerical=data.drop(['ID','make','fuel-type','aspiration','num-of-doors','fuel-system',
                     'body-style','drive-wheel','engine-loc','engine-type','cylinder'],axis=1)
numerical.head()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
#create correlation with hitmap

#create correlation
corr = numerical.corr(method = 'pearson')

#convert correlation to numpy array
mask = np.array(corr)

#to mask the repetitive value for each pair
mask[np.tril_indices_from(mask)] = False
fig, ax = plt.subplots(figsize = (15,12))
fig.set_size_inches(20,10)
sns.heatmap(corr, 
            mask = mask, 
            vmax = 0.9, 
            square = True, 
            annot = True)
```
